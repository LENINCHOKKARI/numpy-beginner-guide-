```python
data = {'A': [1, 2, 2, 3, 11,11],
        'B': [1, 5, 5, 6, 12,25],
}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
q_low = df["B"].quantile(0.25)
q_high = df["B"].quantile(0.75)
```

```python
q_range = q_high - q_low
```

```python
lower_bound = q_low - 1.5 * q_range
upper_bound = q_high + 1.5 * q_range
```

```python
df_filtered = df[(df["A"] >= lower_bound) & (df["A"] <= upper_bound)]
df_filtered
```

group by

```python
data = {'category': ['A', 'B', 'A', "B" , 'A'],
        'Value' : [10,20,30,40,50]}
df = pd.DataFrame(data)
df
```

```python
df.groupby("category").sum()
```

```python
df.groupby("category").mean()
```

```python
df.groupby("category").median()
```

```python
df.groupby("category").agg(["sum","mean","median","min","max"])
```

```python
df.groupby("category")["Value"].sum()
```

```python
df.groupby("category")["Value"].max()
```

```python
df.groupby("category")["Value"].min()
```

```python
custom_agg_func = lambda x : x.max() - x.min()
df.groupby("category")["Value"].agg(custom_agg_func)
```

```python
df.groupby("category")["Value"].agg(["max","min"])
```

```python
def custom_mean(values):
    return sum(values) / len(values)
df.groupby("category")["Value"].agg(custom_mean)
```
